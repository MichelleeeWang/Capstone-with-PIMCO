## Old sample

```python
sampled_result_bysentc = sampled_result.set_index('text')
sampled_result_bysentc
```

```python
sampled_result_bysentc = sampled_result_bysentc[~sampled_result_bysentc.index.duplicated(keep='first')]
```

```python
topics = ['credit', 'fed_funds_rate', 'financial_conditions',
       'financial_markets', 'geopolitical_uncertainty', 'global_growth',
       'growth', 'housing', 'inflation', 'labor_market',
       'liquidity_measures', 'monetary_policy', 'quantitative_easing']
```

```python
topic_dfs = {}
for topic in topics:
    topic_df = pd.read_excel('/content/drive/MyDrive/capstone-pimco/Part1/data/manual_direction_eng/sampled_' + topic + '_direction.xlsx', usecols=[0, 1, 2, 3, 4])
    topic_df['cleaned_text'] = sampled_result_bysentc.loc[list(topic_df['text'].values), :]['cleaned_text'].values
    topic_df['topic'] = topic_df['topic'].astype('str')
    topic_df['new_predicted_topic'] = sampled_result_bysentc.loc[list(topic_df['text'].values), :]['predicted_topics_is_growth'].values
    topic_dfs[topic] = topic_df
    topic_df.to_csv('/content/drive/MyDrive/capstone-pimco/Part1/data/manual_direction_eng_new/sampled_' + topic + '_direction.csv')
```

```python
topics = ['credit', 'fed_funds_rate', 'financial_markets',
       'geopolitical_uncertainty', 'growth', 'housing', 'inflation',
       'labor_market', 'liquidity_measures', 'quantitative_easing']
keys = ['credit', 'fed funds rate', 'financial markets',
       'geopolitical uncertainty', 'growth', 'housing', 'inflation',
       'labor market', 'liquidity measures', 'quantitative easing']
```

```python
for topic in topics:
    topic_df = topic_dfs[topic].copy()
    print(len(topic_df[topic_df['topic'] != topic_df['new_predicted_topic']]))
    topic_df = topic_df[topic_df['topic'] == topic_df['new_predicted_topic']]
    key = topic_df['topic'].values[0]
    leng = len(topic_df)
    for i in range(leng, 100):
        new_sample = sampled_result_singlelabel[sampled_result_singlelabel['predicted_topics_is_growth'] == key].sample(1)
        while new_sample['text'].values[0] in topic_df['text'].values:
            new_sample = sampled_result_singlelabel[sampled_result_singlelabel['predicted_topics_is_growth'] == key].sample(1)
        topic_df = topic_df.append(pd.DataFrame([[
                                       new_sample['uncleaned_text'].values[0],
                                       new_sample['text'].values[0],
                                       key,
                                       np.nan,
                                       None,
                                       new_sample['cleaned_text'].values[0],
                                       key
        ]], columns=topic_df.columns))
    topic_df.to_excel('/content/drive/MyDrive/capstone-pimco/Part1/data/manual_direction_eng_completed/sampled_' + topic + '_direction.xlsx')
```
